```python
initial_auction_table.index = [i + 1 for i in range(count_rows)]
draft_1 = initial_auction_table[(initial_auction_table['LM'] == "L") & (initial_auction_table['BS'] == "B")].sort_values("Price", ascending = True, kind = 'Quicksort')
draft_2 = initial_auction_table[(initial_auction_table['LM'] == "M") & (initial_auction_table['BS'] == "B")]
draft_3 = initial_auction_table[(initial_auction_table['LM'] == "M") & (initial_auction_table['BS'] == "S")]
draft_4 = initial_auction_table[(initial_auction_table['LM'] == "L") & (initial_auction_table['BS'] == "S")].sort_values("Price", ascending = False, kind = 'Quicksort')
```

```python
auction_table = pd.concat([draft_1, draft_2, draft_3, draft_4])
auction_table
```
